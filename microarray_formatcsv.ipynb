```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
labels = pd.read_csv('actual.csv')
print(labels['patient'])
labels.head()
```

```python
labels.cancer.value_counts()
```

```python
train = pd.read_csv('data_set_ALL_AML_train.csv')
print(train.shape)
train.head()
```

```python
test = pd.read_csv('data_set_ALL_AML_independent.csv')
print(test.shape)
test.head()
```

```python
gene_info = train.iloc[:,0:2]
gene_info
```

```python
# function to reformat data
def prepdf(df):
    # get list of 'call' columns
    callcol = [col for col in df.columns if col.startswith('call')]
    # drop 'call' columns
    df = df.drop(callcol, axis=1)
    # set index to gene accession num
    df = df.drop('Gene Description', axis=1).set_index('Gene Accession Number')
    # transpose
    df = df.T
    # create patient col from index (currently not in numerical order)
    df = df.reset_index()
    df.rename(columns={'index':'patient'}, inplace=True)
    # make patient col numeric, put in numeric order, reset index again
    df['patient'] = pd.to_numeric(df['patient'])
    df = df.sort_values(by=['patient']).reset_index(drop=True)
    return df
train = prepdf(train)
train
```

```python
test = prepdf(test)
test
```

```python
# temporarily join train and test dfs
data = pd.concat([train,test], ignore_index=True)
print(data.shape)
data
```

```python
# associate labels with samples
data = data.join(labels.cancer)
data = data.drop('patient', axis=1)
data
```

```python
# resplit into train (first 38 rows), test (last 34 rows)
train = data.loc[0:37,:].copy()
test = data.loc[38:72,:].copy()
```

```python
# check re-split data
print(train.shape)
print(test.shape)
print(train.cancer.value_counts())
print(test.cancer.value_counts())
```

```python
# save formatted dfs as csv files
test.to_csv('microarray_test.csv')
train.to_csv('microarray_train.csv')
gene_info.to_csv('gene_info.csv')
```
